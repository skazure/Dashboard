```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn import linear_model
import statsmodels.api as sm
import tkinter as tk 
 
RmDetails = pd.read_csv(r'C:\Users\sandeep kumar\Documents\GitHub\Dashboard\RMScorePredictionExperiment\RmDetails_Input.csv')
df = DataFrame(RmDetails,columns=['RmPid','Location','Year','TotalKpi'])

#Not creating dummies rather modifying the input data for the simple understanding

plt.scatter(df['RmPid'].astype(float), df['TotalKpi'].astype(float), color='yellow')
plt.title('RmPid Vs TotalKpi', fontsize=14)
plt.xlabel('RmPid', fontsize=14)
plt.ylabel('TotalKpi', fontsize=14)
plt.grid(True)
plt.show()

plt.scatter(df['Location'].astype(float), df['TotalKpi'].astype(float), color='red')
plt.title('Location Vs TotalKpi', fontsize=14)
plt.xlabel('Location', fontsize=14)
plt.ylabel('TotalKpi', fontsize=14)
plt.grid(True)
plt.show()


plt.scatter(df['Year'].astype(float), df['TotalKpi'].astype(float), color='green')
plt.title('Year Vs TotalKpi', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('TotalKpi', fontsize=14)
plt.grid(True)
plt.show()


# here we have 3 variables for multiple regression. 
X = df[['RmPid','Location','Year']].astype(float)
Y = df['TotalKpi'].astype(float)


# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)


print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)


# prediction with sklearn
RmPid = 1
Location = 4
year = 2018
print ('Predicted Kpi for RM:',RmPid,'\n', regr.predict([[RmPid,Location,year]]))



# with statsmodels
X = sm.add_constant(X) # adding a constant
  
model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
  
print_model = model.summary()
print(print_model)


# tkinter GUI
root= tk.Tk() 
  
canvas1 = tk.Canvas(root, width = 1200, height = 450)
canvas1.pack()
 
# with sklearn
Intercept_result = ('Intercept: ', regr.intercept_)
label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center')
canvas1.create_window(260, 220, window=label_Intercept)
 
# with sklearn
Coefficients_result  = ('Coefficients: ', regr.coef_)
label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center')
canvas1.create_window(260, 240, window=label_Coefficients)
 
# with statsmodels
print_model = model.summary()
label_model = tk.Label(root, text=print_model, justify = 'center', relief = 'solid', bg='LightSkyBlue1')
canvas1.create_window(800, 220, window=label_model)
 
# RmPid
label1 = tk.Label(root, text='RmPid (1/2/3/4): ')
canvas1.create_window(100, 100, window=label1)
 
entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)
 
# Location 
label2 = tk.Label(root, text='Location (1/2/3/4): ')
canvas1.create_window(120, 120, window=label2)
 
entry2 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 120, window=entry2)

# Year
label3 = tk.Label(root, text='Year : ')
canvas1.create_window(140, 140, window=label3)
 
entry3 = tk.Entry (root) # create 3rd entry box
canvas1.create_window(270, 140, window=entry3)
 
def values(): 
    global RmPid #our 1st input variable
    RmPid = float(entry1.get()) 
     
    global Location #our 2nd input variable
    Location = float(entry2.get()) 
    
    global Year #our 3rd input variable
    Year = float(entry3.get()) 
     
    Prediction_result  = ('Predicted TotalKpi: ', regr.predict([[RmPid ,Location,Year]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 280, window=label_Prediction)
     
button1 = tk.Button (root, text='Predict Kpi',command=values, bg='orange') 
canvas1.create_window(270, 170, window=button1)
  
 
root.mainloop()

```
